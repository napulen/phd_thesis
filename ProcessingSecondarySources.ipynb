```python
import pandas as pd
import texthero as hero
```

```python
df = pd.read_csv("RNA Timeline - Meta-books.csv")
df["Title_nostopwords"] = hero.remove_stopwords(df["Book"].str.lower())
df["Edition"] = df["Edition"].fillna(1).astype(int)
```

```python
for idx, row in df.iterrows():
    mainauthorid = row['Author'].split(",")[0].lower()
    titleid = row['Title_nostopwords'].split()[0].lower()
    identifier = f"{mainauthorid}{row['Year']}{titleid}"
    bibtex = f"""@book{{{identifier},
    title = {{{row['Book'].title()}}},
    author = {{{row['Author']}}},
    year = {{{row['Year']}}},
    edition = {{{row['Edition']}}}
    keywords = {{PhDThesis_Background_Metabook}}
}}
"""
    print(bibtex)
```
